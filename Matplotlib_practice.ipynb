```
ds=pd.read_csv("dataset wt ht gender.csv")
```

```
hed=ds.head(7)
print(hed)
```

```
#line plot
plt.plot(hed['Height'],hed['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Ht vs Wt')
```

```
#bar chart
ran=ds.sample(10)
#print(ran)
plt.bar(ran['Gender'],height=ran['Height'])
```

```
#pie charts
city = ('Pune', 'Satara', 'Mumbai', 'Kanpur', 'Bhopal', 'Assam')
exp=(0,0.15,0,0,0,0)
rank = [4, 7, 1, 3, 2, 5]
plt.pie(x=rank,explode=exp,labels=city,autopct='%1.1f%%',shadow=True,radius=2)
#plt.pie(x=ran['Height'],labels=ran['Gender'],shadow=True,rotatelabels=True,autopct='%1.1f%%')
```
