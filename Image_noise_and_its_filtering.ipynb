```
impulse_noise = cv2.imread("Uniform random noise.jpg",0)
cv2_imshow(impulse_noise)
ret,impulse_noise = cv2.threshold(impulse_noise,130,255,cv2.THRESH_BINARY)
cv2_imshow(impulse_noise)
cv2.imwrite("Impulse noise.jpg",impulse_noise)
```

```
gaussian_noise=(gaussian_noise*0.5).astype(np.uint8)
noisy_pup1=cv2.add(pup,gaussian_noise)
cv2_imshow(pup)
cv2_imshow(noisy_pup1)
cv2.imwrite("Noisy_pup1.jpg",noisy_pup1)
```

```
uniform_noise=uniform_noise.astype(np.uint8)
noisy_pup2=cv2.add(pup,uniform_noise)
cv2_imshow(pup)
cv2_imshow(noisy_pup2)
cv2.imwrite("Noisy pup 2.jpg",noisy_pup2)
```

```
impulse_noise=impulse_noise.astype(np.uint8)
noisy_pup3=cv2.add(pup,impulse_noise)
cv2_imshow(pup)
cv2_imshow(noisy_pup3)
cv2.imwrite("Noisy pup 3.jpg",noisy_pup3)
```

```
blur1=cv2.medianBlur(noisy_pup1,3)
cv2_imshow(blur1)
cv2.imwrite("Median Blur pup 1.jpg",blur1)
```

```
blur2=cv2.blur(noisy_pup1,(5,5))
cv2_imshow(blur2)
#cv2.write("Average blur pup 1.jpg",blur2)
```

```
blur3=cv2.GaussianBlur(noisy_pup1,(3,3),1)
cv2_imshow(blur3)
                       
```

```
bi_pup=cv2.bilateralFilter(noisy_pup1,15,75,75)
cv2_imshow(bi_pup)
```

```
ker=np.ones((5,5),np.float32)/25
tod=cv2.filter2D(noisy_pup1,-1,ker)
cv2_imshow(tod)
```
