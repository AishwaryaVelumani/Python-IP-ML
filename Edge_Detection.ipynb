```
img=cv2.imread('mouse.png',cv2.IMREAD_GRAYSCALE)
```

```
from google.colab.patches import cv2_imshow
```

```
img1=cv2.GaussianBlur(img,(3,3),0)
plt.subplot(1,2,1),plt.imshow(img,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(1,2,2),plt.imshow(img1,cmap='gray')
plt.title('Blur'),plt.xticks([]),plt.yticks([])
```

# Sobel Edge Detector

```
sobelx=cv2.Sobel(img1,cv2.CV_64F,1,0)
cv2_imshow(sobelx)
print()
sobely=cv2.Sobel(img1,cv2.CV_64F,0,1)
cv2_imshow(sobely)
```

# Laplacian Edge Detector

```
lap=cv2.Laplacian(img1,cv2.CV_64F,ksize=5)
cv2_imshow(lap)
```

# Canny Edge Detector

```
can=cv2.Canny(img1,100,150)
cv2_imshow(can)
```

# Morphological Operations

```
from google.colab import files
uploaded=files.upload()
```

```
img=cv2.imread('IO.png',0)
(ht,wd)=img.shape
print(ht,' ',wd)
```

```
ker=np.ones((5,5),np.uint8)
```

```
max_int=255
out=img.copy()
for i in np.arange(ht):
    for j in np.arange(wd):
        a=out.item(i,j)
        b=max_int-a
        out.itemset((i,j),b)

r,bwimg=cv2.threshold(out,24,255,cv2.THRESH_BINARY)
cv2_imshow(bwimg)
```

```
img_ero=cv2.erode(bwimg,ker,iterations=2)
img_dil=cv2.dilate(bwimg,ker,iterations=2)
print("Original")
print()
cv2_imshow(bwimg)
print("Eroded")
print()
cv2_imshow(img_ero)
print("Dilated")
cv2_imshow(img_dil)
```
