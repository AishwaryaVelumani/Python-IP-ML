```
ds=pd.read_csv("iris1.csv")
#scatter plot
res=sns.scatterplot(x='petal.length',y='variety',data=ds)
```

```
tip=sns.load_dataset("tips")
print(tip)
```

```
#cat plot
res=sns.catplot(x='total_bill',y='smoker',jitter=True,data=tip)
```

```
#strip plot
res=sns.stripplot(x='total_bill',y='smoker',data=tip,jitter=True)
```

```
res=sns.swarmplot(x='total_bill',y='smoker',data=tip)
```

```
#box plot
res=sns.boxplot(x='total_bill',y='sex',data=tip)
```

```
#violin plot
sns.violinplot(x='total_bill',y='sex',data=tip)
```

```
#boxen plot
sns.boxenplot(x='total_bill',y='sex',data=tip)
```

```
#count plot
res=sns.countplot(y='total_bill',data=tip)
```

```
res=sns.barplot(y='tip',data=tip)
```

```
res=sns.pointplot(x='size',data=tip)
```

```
sns.set_style('darkgrid')
res=sns.distplot(tip['tip'])
```

```
res=sns.jointplot(x='total_bill',y='tip',data=tip)
```
