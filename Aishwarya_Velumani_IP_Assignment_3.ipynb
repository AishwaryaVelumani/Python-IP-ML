```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded= files.upload()
```

```
img=cv2.imread("sia2.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```
roi= img[0:270,37:235]
cv2_imshow(roi)
```

```
import imutils
rotated=imutils.rotate(img,-45)
cv2_imshow(rotated)
```

```
rotated1=imutils.rotate_bound(img,+45)
cv2_imshow(rotated1)
```

```
img1=np.flipud(img)
cv2_imshow(img1)
```

```
blue,green,red=cv2.split(img)
cv2_imshow(blue)
cv2_imshow(green)
cv2_imshow(red)
```

```
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
print("Size:",str(img.size))
```

```
(h,w,d)= img.shape
print("width={}, height={}, depth={}".format(w,h,d))
```

```
max_intensity=255
gray1=gray.copy()
for i in np.arange(h):
    for j in np.arange(w):
        a=gray1.item(i,j)
        b=max_intensity - a
        gray1.itemset((i,j),b)
cv2.imwrite('images/brightness.jpg',gray1)
cv2_imshow(gray1)
```

```
gray2=gray.copy()
ret,thresh = cv2.threshold(gray2, 125, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
```
