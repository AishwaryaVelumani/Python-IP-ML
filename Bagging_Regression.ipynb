```
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```
x,y= make_regression(n_samples=1000,n_features=20,n_informative=15,noise=0.1,random_state=5)
print(x.shape,y.shape)
```

```
dtr=DecisionTreeRegressor()
bgr= BaggingRegressor(base_estimator=dtr,n_estimators=20,n_jobs=3,random_state=5)
bgr.fit(x,y)
ypr=bgr.predict(x)
```

```
print(ypr.shape)
print(y.shape)
```

```
#accuracy_score is only for classification
print('Score: ',100*bgr.score(x,y))
```

**Visualisation**

```
plt.figure(figsize=(15,8))

plt.subplot(121)
plt.scatter(x[:,0],y,c='green',label='Original Data Points')
plt.legend()

plt.subplot(122)
plt.scatter(x[:,0],ypr,c='blue',label='Predicted Data Points')
plt.legend()
```
